```python
# патерн проектирования

class Unit:
    def move(self, star,level=0):
        print(f'{" " * level}{self} направляется к звезде{star} ')

    def shoot(self, level=0):
        print(f'{" " * level}{self} аттакует прешельцев')  # исправлено: {" " * level}


class spaceship(Unit):
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f'космический корабль {self.name}'


class SpaceFleet(Unit):
    def __init__(self, name, admiral):
        self.name = name
        self.admiral = admiral
        self.units = []

    def append(self, unit):
        self.units.append(unit)

    def __str__(self):
        return f'космический флот {self.name} под командованием {self.admiral}'

    def shoot(self, level=0):
        super().shoot(level)  # вызываем родительский shoot с уровнем
        for unit in self.units:
            unit.shoot(level + 1)  # вызываем shoot у каждого юнита с увеличенным уровнем

    def move(self,star,level=0):
        super().move(star)
        for unit in self.units:
            unit.move(star,level+1)


galaxy = spaceship('galaxy')
# galaxy.move(" Андромеда")?
# galaxy.shoot()
terrafleet = SpaceFleet("Земной Флот",'Данил')
terrafleet.append(galaxy)
terrafleet.append(spaceship("Enterprise"))

marsfleet = SpaceFleet("Марсианский флот", 'Рома')
marsfleet.append(spaceship("Биба"))
marsfleet.append(spaceship("Боба"))

unitedfleet  = SpaceFleet("Обьедененный флот звездной федерации",'Пашка')
unitedfleet.append(spaceship("Мини-помощник"))
unitedfleet.append(marsfleet)
unitedfleet.append(terrafleet)
unitedfleet.move("Эльдебарано")
unitedfleet.shoot()
```
