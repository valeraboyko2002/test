```python
import pandas as pd 

df = pd.read_csv("csv/titanic.csv")
df
```

```python
df.drop(['Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
target=df.Survived
inp = df.drop('Survived',axis='columns')
```

```python
dum = pd.get_dummies(inp.Sex)
dum.head()
inp = pd.concat([inp,dum],axis='columns')
inp.drop('Sex',axis='columns',inplace=True)
```

```python
inp.head(3)
```

```python
inp.columns[inp.isna().any()]
```

```python
inp.Age = inp.Age.fillna(inp.Age.mean())
inp.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inp, target, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
X_test.head()
```

```python
X_test.head()
```

```python
model.predict(X_test[:10])
```

```python
model.predict_proba(X_test)
```

```python
df = pd.read_csv('csv/spam.csv')
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x == 'spam' else 0)
```

```python
df.drop('Category',axis='columns',inplace = True)
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df.Message,df.spam,test_size=0.25)
```

```python
# данная хуйня берет слова которые повторяются бОльшее кол-во раз. 
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()
```

```python
# помещаем значения обучающей выборки в модель после чего трансформируем его в вектора

X_train_count = vect.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey dude whatsupp how are u doing',
    'Upto 20% discount in our store ! weekends are free food. join to our partnership'
]
email_count = vect.transform(emails)
model.predict(email_count)
```

```python
x_test_count = vect.transform(X_test.values)
model.score(x_test_count, y_test)
```

```python
from sklearn.pipeline import Pipeline

clf=Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```
