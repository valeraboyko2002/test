```python
# ONE HOT ENCODING 
df = pd.read_csv("csv/carprices.csv")
df = df.rename(columns={"Car Model": "model", "Sell Price($)": "price",'Age(yrs)':'age'})
df
```

```python
from sklearn.preprocessing import LabelEncoder
de = LabelEncoder()
df.model = de.fit_transform(df.model)
df
```

```python
X = df.drop("price",axis="columns").values
y = df.price
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

tranform = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(), [0]) # [0] - индекс столбца с категориями
        ],
        remainder='passthrough')

x_encoded = tranform.fit_transform(X)
```

```python
x_encoded
```

```python
X = x_encoded[:,1:]
X
```

```python
model = LinearRegression()
model.fit(X,y)
model.score(X,y)
```
