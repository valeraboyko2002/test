```python
import pandas as pd 
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==0].head() # setosa
df[df.target==1].head() # versicolor
df[df.target==2].head() # virginica
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['target','flower_name'],axis = 'columns')
X.head()
```

```python
y= df.target
```

```python
X_train,X_test,y_train, y_test = train_test_split(X,y,random_state=1,test_size=0.2)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
# HOMEWORK
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

digits = load_digits()
```

```python
X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target,random_state = 1, test_size=0.8)
model = SVC(kernel='sigmoid')
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
model = SVC(kernel='linear')
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
model = SVC(kernel='rbf')
model.fit(X_train, y_train)
model.score(X_test,y_test)
```

```python
model = SVC(C=20,kernel='rbf')
model.fit(X_train, y_train)
model.score(X_test,y_test)
```
